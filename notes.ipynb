## 📦 3. Why the `src` Layout is Used

❌ **Bad structure**

```
project/
   nexusview/
```

✅ **Correct structure**

```
project/
   src/
      nexusview/
```

**Reason**

- Prevent accidental imports
- Ensure package works correctly after `pip install`

## 📁 7. Folder Creation

```python
os.makedirs(path, exist_ok=True)
```

Meaning of `exist_ok=True`

- If folder exists → **do nothing**
- If folder doesn't exist → **create it**

Prevents errors.

---

## 📄 8. File Creation Logic

Script checks:

- If **file does not exist**
- OR **file is empty**

Then create file.

If file already contains code:

⚠ **Do NOT overwrite file**
