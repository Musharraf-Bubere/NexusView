# 📁 Why `__init__.py` is Required

Example:

```
mypackage/
   __init__.py
   module.py
```

Purpose:

- Marks folder as **Python package**
- Allows imports
- Can run initialization code

Example:

```python
import mypackage.module
```

Without `__init__.py`, Python may not recognize the folder as a package.
